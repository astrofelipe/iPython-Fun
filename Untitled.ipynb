```python
from linfit import linfit
import numpy as np
import emcee
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina' 
```

```python
realm = 1.5
realb = -2.0

N = 50

x     = np.sort(5*np.random.rand(N))
err_y = np.random.uniform(1, 5, size=N)
y     = realm*x + realb + np.random.normal(0, err_y, size=N)

(lm, lb), lcov = linfit(x, y)
print lm, lb
print lcov

fig, ax = plt.subplots(figsize=[8,3])
ax.errorbar(x, y, yerr=err_y, fmt='ok')
ax.plot(x, lm*x+lb, '-r')
ax.plot(x, realm*x+realb, '-g')
```

```python
def ln_prior(theta):
    m, b = theta
    if -3.0 < m < 3.0 and -3.0 < b < -3.0:
        return 0.0
    return 0.0

def ln_likelihood(theta, x, y, err_y):
    m, b  = theta
    chi   = (m*x + b - y) / err_y
    return -0.5*np.sum(chi**2)

def ln_posterior(theta, x, y, err_y):
    return ln_prior(theta) + ln_likelihood(theta, x, y, err_y)
```

```python
nwalkers = 4
ndim     = 2
sampler  = emcee.EnsembleSampler(nwalkers=nwalkers, dim=ndim, lnpostfn=ln_posterior, args=(x, y, err_y))

p0 = np.random.uniform(-3., 3, size=(nwalkers,ndim))

#Burn-in
pos, prob, state = sampler.run_mcmc(p0, 1000)

samples = sampler.chain[:, 100:, :].reshape((-1, ndim))


fig, ax = plt.subplots(nrows=2, figsize=[10,4])
for chain in sampler.chain:
    ax[0].plot(chain[:,0], '-k')
    ax[1].plot(chain[:,1], '-k')

print np.median(sampler.flatchain, axis=0)

import corner
fig = corner.corner(samples, labels=["$m$", "$b$"],
                      truths=[realm, realb])
```
